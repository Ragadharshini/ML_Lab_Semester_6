```python
import pandas as pd
import numpy as np 

teleco_df=pd.read_csv('Telco-Customer-Churn.csv')

teleco_df
```

```python
teleco_df.drop(['customerID'],axis = 1,inplace = True)

teleco_df
```

```python
from sklearn import preprocessing 

obj_1 = list(teleco_df.select_dtypes(include=['object']).columns)
teleco_df[obj_1]

label_encoder = preprocessing.LabelEncoder() 

for column in obj_1:
    
    teleco_df[column] = label_encoder.fit_transform(teleco_df[column])

teleco_df.head()
```

```python
teleco_df['TotalCharges'] = teleco_df['TotalCharges'].fillna(0)

teleco_df
```

```python
from sklearn.preprocessing import StandardScaler

teleco_df_1 = StandardScaler()
teleco_df_1.fit(teleco_df.drop(['Churn'],axis = 1))

attributes = teleco_df_1.transform(teleco_df.drop('Churn',axis = 1))

teleco_df_1
```

```python
from sklearn.model_selection import train_test_split

X = attributes

Y = teleco_df['Churn']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=27)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score ,classification_report, confusion_matrix

model1 = LogisticRegression()

model1.fit(X_train,Y_train)

y1_predicted = model1.predict(X_test)
y1_predicted
```

```python
print(classification_report(Y_test,y1_predicted))
```

```python
accuracy_score(Y_test, y1_predicted)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier()

dt_classifier.fit(X_train, Y_train)

y2_predicted = dt_classifier.predict(X_test)

y2_predicted
```

```python
print(classification_report(Y_test,y2_predicted))
```

```python
accuracy_score(Y_test,y2_predicted)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, Y_train)

y3_predicted = svc.predict(X_test)
```

```python
y3_predicted
```

```python
print(classification_report(Y_test,y3_predicted))
```

```python
accuracy_score(Y_test,y3_predicted)
```

```python
from sklearn.naive_bayes import GaussianNB

naive_bayes_classifier = GaussianNB()

naive_bayes_classifier.fit(X_train, Y_train)
```

```python
y4_predicted=naive_bayes_classifier.predict(X_test)

y4_predicted
```

```python
print(classification_report(Y_test,y4_predicted))
```

```python
accuracy_score(Y_test,y4_predicted)
```

```python
from sklearn.neighbors import KNeighborsClassifier

custom_knnclassifier = KNeighborsClassifier(n_neighbors = 30)

custom_knnclassifier.fit(X_train,Y_train)
```

```python
y5_predicted = custom_knnclassifier.predict(X_test)
```

```python
print(accuracy_score(Y_test,y5_predicted))
```
